```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Bank%20Churn%20Modelling.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
df.drop(['CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
df
```

```python
df.info()
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
df.Churn.value_counts()
```

```python
y = df['Churn']
X = df.drop(['Churn'],axis=1)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
```

```python
X.shape,y.shape,y.value_counts()
```

```python
X_ros,y_ros = ros.fit_resample(X,y)
```

```python
X_ros.shape,y_ros.shape,y_ros.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_ros,y_ros,random_state=2529)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_predict))
```
